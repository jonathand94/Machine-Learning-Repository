## 3. Funciones y ciclos

### Funciones

```python
# Creando una función
import math

def calcular_area_circulo(r):
    """
        params: 
            r: [double] non-negative value indicating the radius of the circle
        return: 
            A: [double] area of the cricle
    """
    if r <= 0:
        raise ValueError('Not a valid radius, please provide a positive radius')
    return math.pi*r**2
```

```python
# Analizando la función print
r = 1
print('ÁREA DEL CÍRCULO CON RADIO DE {0} = {1}'.format(r, calcular_area_circulo(r)))
```

```python
# Regresando errores
calcular_area_circulo(0)
```

## 4. Numpy, Matplotlib, Pillow, SciPy

### Numpy

```python
import numpy as np
```

```python
# Creando matriz de 2x2
A = np.array([[4, 1], [-6, -3]])
print(A)
```

```python
# Calcular el determinante

detA = np.linalg.det(A)
print(detA)
```

```python
# Obteniendo los valores y vectores propios
e = np.linalg.eig(A)
e_values = e[0]
e_vectors_norm = e[1]

print('Valores propios: {0} \n'.format(e_values))
print('Vectores propios: {0} \n'.format(e_vectors_norm))
```

```python
# Multiplicación matricial - rotacion
import math

def rotar_vector2D(x, angle):
    """
        params: 
            x: numpy array of dimensions (1 x 2)
            angle: angle in degreed for which the vector x will be rotated
        return: 
            x_rot: numpy array (1 x 2) indicating the rotated vector
    """
    # Transformar de grados a radianes
    angle = angle * (math.pi) / 180
    # Matriz de rotación
    A = np.array([[math.cos(angle), -math.sin(angle)], [math.sin(angle), math.cos(angle)]])
    
    return np.matmul(A, x)
```

```python
# Proponer vector x = [1, 0] y rotarlo 90°
x = np.array([[1, 0]])
x = np.transpose(x)
x_rot_1 = rotar_vector2D(x, 90)
x_rot_2 = np.rot90(x)
print(x_rot_1)
print(x_rot_2)
```

```python
# Resolver sistemas de ecuaciones
b = [1, 2, 3, 4]
A = [[1, 5, -2, 7], [-1, 1, 8, 2], [1, 5, -3, 15], [1, 2, 5, -1]]
x = np.linalg.solve(A, b)
print(x)
```

### Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Crear vector x
x = np.linspace(-2*math.pi, 2*math.pi)
print(x, x.shape)
```

```python
# Función
y = np.sin(x)
print(y)
```

```python
# Graficar
plt.figure(1)
plt.plot(x, y)
```

```python
# Calculamos la derivada
y_prime = np.gradient(y)
print(y_prime)
```

```python
# Graficamos ambas curvas
plt.figure(1)
plt.plot(x, y, color='r')
plt.plot(x, 4*y_prime, color='b')
```
